```python
import pandas as pd
import json
import yaml
import os
import pymongo
from pymongo import MongoClient
```

```python
file = "yelp_academic_dataset_business.json"
list = []
```

```python
review = pd.read_json(file, lines=True)
review
```

```python
yelp_vegas = review[review['city']=="Las Vegas"]
yelp_vegas.head()    
```

```python
open_vegas = yelp_vegas[yelp_vegas['is_open']==1]
open_vegas.head()
```

```python
len(open_vegas)
```

```python
vegasDict = open_vegas.to_dict('records')
vegasDict
```

```python
conn = (f'mongodb://localhost:27017/')
client = pymongo.MongoClient(conn)
database_name = 'yelpBusiness'
collection_name = 'openVegas'
```

```python
db = client.yelpBusiness
collection = db.openVegas.find()
```

```python
db.openVegas.insert_many(vegasDict)
```
